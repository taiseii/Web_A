```python
def PageRankCalc(file):
    # load the edge list
    edgelist = []
    with open(file) as f:
        for line in f:
            if not line.startswith('%'):
                nodes = line.split()
                edgelist.append((int(nodes[0]), int(nodes[1])))

    # create a directed Graph using Networkx
    G = nx.DiGraph()
    G.add_edges_from(edgelist)

    # calculate PageRank directly use
    # the function from NetworkX
    # more details about this function
    # can be found 
    # https://networkx.github.io/documentation/latest/reference/algorithms/generated/networkx.algorithms.link_analysis.pagerank_alg.pagerank.html
    pr = nx.pagerank(G, alpha=0.85)
    # the return value pr is dict in Python
    # in whici the keys are the node indices 
    # and the values are corresponding PagRank values
#     print(pr)
    return pr
```

```python
def LouvainAlgorithm(file):
    # load the edge list and create a directed Graph
    # load the edge list
    edgelist = []
    with open(file) as f:
        for line in f:
            if not line.startswith('%'):
                nodes = line.split()
                edgelist.append((int(nodes[0]), int(nodes[1])))

    # create a Graph using Networkx
    # note that we create an undirected graph for simplicity
    G = nx.Graph()
    G.add_edges_from(edgelist)

    # This method implements Louvain method using a greedy algorithm
    # This is the partition of highest modularity, 
    # i.e. the highest partition of the dendrogram generated by the Louvain algorithm
    # more details: http://perso.crans.org/aynaud/communities/api.html#community.best_partition
    partition = community.best_partition(G)

    # partition is the returned results of best_partition
    # it is a dict in Python
#     print(partition)
    return partition
```

<h3>Creating original graph G0</h3>

```python
def make_G0():
    G0 = nx.DiGraph()
    edgelist0 = []
    with open('out.advogato') as f:
        for line in f:
            if not line.startswith('%'):
                nodes = line.split()
                edgelist0.append((int(nodes[0]), int(nodes[1])))
        G0.add_edges_from(edgelist0)
    return G0
```

Above function is used to create given graph.<br>
It is called before edge/nodes and corresponding edges are removed in various manners indicated in the assignment description.<br>

```python
# remove uniform number of edges 
rm_n = 10000
```

Above is variable used to perform uniformly at random for both nodes and edges

<h3>edge uniform</h3> <br>
Below is used to remove edge uniformly at random from the original graph which has been provided. At most there will be 10 iteration in order to reduce the complexitiy of computing measurements to compare PageRank etc.<br>
In this scenario, after the initialization of the original graph, operation below is going to remove rm_n number of edges from garph by randomly picking the edge to be removed.<br>
This produce 'edge_uniform.csv'

```python
G1=make_G0()
for i in range(10):
    num_iter = str(len(list(G1.edges)))
    pr0 = nx.pagerank(G1, alpha=0.85)
    pd.DataFrame.from_dict(data=pr0,
                           orient='index').to_csv(
        num_iter+'_edge_uniform_'+'.csv', header=False)
    for rm in range(rm_n):
        index = randint(1, len(list(G1.edges))-1)
        G1.remove_edge(list(G1.edges)[index][0], list(G1.edges)[index][1])
```

<h3>node uniform</h3>
Below is used to remove node uniformly at random from the original graph which has been provided. At most there will be 10 iteration in order to reduce the complexitiy of computing measurements to compare PageRank etc.
<br>
When a node is removed from the graph its corresponding edge is also going to be removed simultaneously.<br>
In this scenario, after the initialization of the original graph, operation below is going to remove rm_n number of nodes from garph by randomly picking the node to be removed.<br>
This produce 'node_uniform.csv'

```python
G2=make_G0()
for i in range(10):
    num_iter = str(len(list(G2.nodes)))
    pr0 = nx.pagerank(G2, alpha=0.85)
    pd.DataFrame.from_dict(data=pr0,
                           orient='index').to_csv(
        num_iter+'_node_uniform_'+'.csv', header=False)
    for rm in range(rm_n):
#         print((len(G2.nodes)-1))
        if len(G2.nodes) != 0:
            index = randint(0, (len(G2.nodes)-1))
            G2.remove_node(list(G2.nodes)[index])
```

<h3>nodes random but proportional</h3>
Below is the operation to remove nodes randomly but proportional to the node degree.<br>
In this scenario, above have been achieved by firstly, obtaining degree of all the nodes inside the graph and assigning all the nodes to the corresponding number of degrees.<br>
Secondly, by categorizing the nodes according to the degree we will decide how many of nodes we would like to remove from list of nodes(This is percent_rm = math.ceil(0.1*j* len(degree_list) )).<br>
Thirdly, we have iterated percent_rm number of times to remove random node from the nodes categorized according to the degree.<br>
This produce 'nodes_proportional.csv'

```python
G3 = make_G0()
for j in range(10):
    degsum = pd.DataFrame()
    indexz = []
    degreez = []
    alredy_removed = False
    for i in range( len(list(G3.nodes)) ):
        try:
            G3.degree(i)
            indexz.append(i)
            degreez.append(G3.degree(i))
        except:
            already_removed = True
    
    degsum['indexz'] = indexz
    degsum['degreez'] = degreez
    
    deg_df = degsum.groupby('degreez')['indexz'].unique()
    
    num_iter = str(len(list(G3.edges)))
    pr0 = nx.pagerank(G3, alpha=0.85)
    pd.DataFrame.from_dict(data=pr0, orient='index').to_csv(
        num_iter+'_proportional_'+'.csv', header=False)
        
    for uq_elm in degsum['degreez'].unique():
        degree_list = list(deg_df[uq_elm])
        percent_rm = math.ceil(0.1*j* len(degree_list) )
        for rm in range(percent_rm):
                node_number = np.random.choice(list(degree_list), 1)
                degree_list.remove(node_number[0])    
                G3.remove_node(node_number[0])
```

<h3>edges random but proportional</h3>

Below is the operation to remove edge randomly but proportional to the node degree.<br>
In this scenario, above have been achieved by firstly, obtaining degree of all the nodes inside the graph and assigning all the nodes to the corresponding number of degrees.<br>
Secondly, by categorizing the nodes according to the degree we will decide how many percent of edge will be removed from the indivisual node(percent_rm = math.ceil(percent*j* uq_elm )).<br>
Thirdly, we have iterated percent_rm number of times to remove random edge from the nodes categorized according to the degree. In order to do this we have decided to iterate edges in the node and using binomial distribution(bin_percent variable below is success parameter) to decide removal of the particular edge of the node<br>
This produce 'edge_proportional.csv'

```python
percent = 0.7
bin_percent = 0.6
G4 = make_G0()
for j in range(10):
    degsum = pd.DataFrame()
    indexz = []
    degreez = []
    alredy_removed = False
    for i in range( len(list(G4.nodes)) ):
        try:
            G4.degree(i)
            indexz.append(i)
            degreez.append(G4.degree(i))
        except:
            alredy_removed = True
    
    degsum['indexz'] = indexz
    degsum['degreez'] = degreez
    
    deg_df = degsum.groupby('degreez')['indexz'].unique()
    
    num_iter = str(len(list(G4.edges)))
    pr0 = nx.pagerank(G4, alpha=0.85)
    pd.DataFrame.from_dict(data=pr0, orient='index').to_csv(num_iter+'_proportional_edge'+'.csv', header=False)
        
    for uq_elm in degsum['degreez'].unique():
        degree_list = list(deg_df[uq_elm])
#         print(degree_list)
        percent_rm = math.ceil(percent*j* uq_elm )
        for rm in range(percent_rm):
                node_number = np.random.choice(list(degree_list), 1)
                edge_of_node = G4.edges(node_number)
                for edges in list(edge_of_node):
                    u = edges[0]
                    v = edges[1]
                    
                    binomial = np.random.binomial(1,bin_percent)
                    if binomial == 1:
#                         degree_list.remove(node_number[0]) 
                        G4.remove_edge(u,v)
                        
```
